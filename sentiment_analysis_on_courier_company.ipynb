```python
# Changes in rabin karp algorithm : 
# No need to loop across the text as the length of pattern and text is same
# Therefore, i have remove the for loop : for i in range(N-M+1) and also calculation for hash value of trailing digit and leading digit and the h value.    
def search(pat, txt, q):
    M = len(pat)
    N = len(txt)

    if (M != N):
      return None
      
    p = 0  
    t = 0 
    d = 256 

    for i in range(M):
        p = (d * p + ord(pat[i])) % q
        t = (d * t + ord(txt[i])) % q
    
    if p == t:
      for j in range(M):
          if txt[j] != pat[j]:
              break
      j += 1
      if j == M:
        return pat
```
